```python
import re
from collections import Counter
import matplotlib.pyplot as plt
```

```python
with open("english.txt", 'r') as f:
  english_sentences = [line.strip() for line in f.readlines()]
```

```python
print(english_sentences[:20])
```

```python
with open("sinhala.txt", 'r') as f:
  sinhala_sentences = [line.strip() for line in f.readlines()]
```

```python
print(sinhala_sentences[:20])
```

```python
# tokenization and preprocessing function
def preprocess(sentence):
  sentence = sentence.lower()
  sentence = re.sub(r'[^\w\s]', '', sentence)
  words = sentence.split()
  return words
```

```python
english_words = []
for sentence in english_sentences:
  words = preprocess(sentence)
  english_words.extend(words)
```

```python
print(english_words[:20])
print(len(english_words))
```

```python
sinhala_words = []
punctuation_pattern = re.compile(r'[!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]')
for sentence in sinhala_sentences:
  sentence = re.sub(r'[^\u0D80-\u0DFF\s]', '', sentence)
  words = sentence.split()
  sinhala_words.extend(words)
```

```python
print(sinhala_words[:20])
print(len(sinhala_words))
```

```python
english_word_freq = Counter(english_words)
sinhala_word_freq = Counter(sinhala_words)
```

```python
print(english_word_freq)
```

```python
print(sinhala_word_freq)
```

```python
english_sorted_freq = english_word_freq.most_common(100)
sinhala_sorted_freq = sinhala_word_freq.most_common(100)
```

```python
print(english_sorted_freq[:30])
```

```python
print(sinhala_sorted_freq[:10])
```

```python
english_sorted_words, english_words_count = zip(*english_word_freq.most_common())
```

```python
import matplotlib.pyplot as plt

# Choose the interval
interval = 5

# Select the words and counts at the specified intervals
selected_indices = range(0, len(english_sorted_words), interval)
selected_words = [english_sorted_words[i] for i in selected_indices]
selected_counts = [english_words_count[i] for i in selected_indices]

# Plot the selected words
plt.figure(figsize=(14, 7))
plt.bar(selected_words, selected_counts, color='blue')
plt.xlabel('English Words')
plt.ylabel('Frequency')
plt.title('English Word Frequencies (Selected Every {} Words)'.format(interval))
plt.xticks(rotation=90)
plt.show()
```

```python
sinhala_sorted_words, sinhala_words_count = zip(*sinhala_word_freq.most_common())
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Choose the interval
interval = 5

# Select the words and counts at the specified intervals
selected_indices = range(0, len(sinhala_sorted_words), interval)
selected_sinhala_words = [sinhala_sorted_words[i] for i in selected_indices]
selected_sinhala_counts = [sinhala_words_count[i] for i in selected_indices]

# Plot the selected words
plt.figure(figsize=(14, 7))
plt.bar(selected_sinhala_words, selected_sinhala_counts, color='green')
plt.title('Sinhala Word Frequencies (Selected Every {} Words)'.format(interval))
plt.xlabel('Sinhala Words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# English ranks
english_ranks = {}
current_rank = 1

for rank, (word, freq) in enumerate(english_sorted_freq):
    if rank == 0:
        # Assign the first rank
        english_ranks[word] = current_rank
    else:
        # If the current word has the same frequency as the previous word, assign the same rank
        if english_sorted_freq[rank][1] == english_sorted_freq[rank - 1][1]:
            english_ranks[word] = current_rank
        else:
            # Increment the rank by 1
            current_rank += 1
            english_ranks[word] = current_rank
```

```python
print(english_ranks)
```

```python
print(list(english_ranks.items())[:15])
```

```python
# Sinhala ranks
sinhala_ranks = {}
current_rank = 1

for rank, (word, freq) in enumerate(sinhala_sorted_freq):
    if rank == 0:
        # Assign the first rank
        sinhala_ranks[word] = current_rank
    else:
        # If the current word has the same frequency as the previous word, assign the same rank
        if sinhala_sorted_freq[rank][1] == sinhala_sorted_freq[rank - 1][1]:
            sinhala_ranks[word] = current_rank
        else:
            # Increment the rank by 1
            current_rank += 1
            sinhala_ranks[word] = current_rank
```

```python
print(list(sinhala_ranks.items())[:15])
```

```python
# calculate f * r for the first 25 words
def zipfs_law(freq_list, rank_dict):
  zipf_products = []
  for word, freq in freq_list[:25]:
    rank = rank_dict[word]
    zipf_product = freq * rank
    zipf_products.append(zipf_product)
  return zipf_products
```

```python
english_zipf_products = zipfs_law(english_sorted_freq, english_ranks)
```

```python
print(english_zipf_products)
```

```python
sinhala_zipf_products = zipfs_law(sinhala_sorted_freq, sinhala_ranks)
```

```python
print(sinhala_zipf_products)
```

```python
# Check if they are approximately constant
plt.figure(figsize=(10, 6))
plt.plot(english_zipf_products, label='English', color='blue', marker='o')
plt.plot(sinhala_zipf_products, label='Sinhala', color='green', marker='o')
plt.title('Zipf\'s Law: f * r Product for Top 25 Words')
plt.xlabel('Word Rank')
plt.ylabel('f * r Product')
plt.legend()
plt.show()
```
